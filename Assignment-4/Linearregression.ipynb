```python
X=df[['crim','zn','indus','chas','nox','rm','age','dis','rad','tax','ptratio','b','lstat']]
Y=df['medv']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.4,random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
Y_train.shape
```

```python
Y_test.shape
```

```python
lrobject=LinearRegression()
```

```python
lrobject.fit(X_train,Y_train)
```

```python
Y_prediction=lrobject.predict(X_test)
```

```python
Y_prediction
```

```python
from sklearn import metrics
```

```python
metrics.mean_absolute_error(Y_test,Y_prediction)
```

```python
metrics.mean_squared_error(Y_test,Y_prediction)
```

```python
np.sqrt(metrics.mean_squared_error(Y_test,Y_prediction))
```
