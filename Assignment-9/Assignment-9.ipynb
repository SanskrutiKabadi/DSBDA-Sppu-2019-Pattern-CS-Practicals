```python
sns.boxplot(x='Age',data=df)
```

```python
Q1=np.percentile(df['Age'],25,interpolation='midpoint')
```

```python
print(Q1)
```

```python
Q3=np.percentile(df['Age'],75,interpolation='midpoint')
```

```python
print(Q3)
```

```python
IQR=Q3-Q1
```

```python
print(IQR)
```

```python
df.shape
```

```python
upper=np.where(df['Age']>=Q3+0.1*IQR)
```

```python
lower=np.where(df['Age']<=Q1-0.1*IQR)
```

```python
df=df.drop(lower[0])
```

```python
df=df.drop(upper[0])
```

```python
df.shape
```

```python
sns.boxplot(x='Age',data=df)
```
